```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from starkhelium import *
from tqdm import trange, tqdm
from scipy.constants import h, hbar, c, alpha, m_e, e, epsilon_0, atomic_mass, pi, physical_constants
a_0 = physical_constants['Bohr radius'][0]

En_h = alpha**2.0 * m_e * c**2.0;
E_h = 4.35974465054e-18;
scl = c*10**-9 * En_h /(h * c);
```

## E = 0.84 V/cm

```python
data_73_73 = np.load('data/StarkMapData_n_73-73_E_0-84_0-84_1_B_1-5776.npy')
data_72_73 = np.load('data/StarkMapData_n_72-73_E_0-84_0-84_1_B_1-5776.npy')
data_71_74 = np.load('data/StarkMapData_n_71-74_E_0-84_0-84_1_B_1-5776.npy')
data_70_75 = np.load('data/StarkMapData_n_70-75_E_0-84_0-84_1_B_1-5776.npy')
```

```python
state_70_73s = 70**2 + 71**2 + 72**2 + 1
state_71_73s = 71**2 + 72**2 + 1
state_72_73s = 72**2 + 1
state_73_73s = 1
energy_73 = []
energy_73.append(data_73_73[0,state_73_73s-1]*scl)
energy_73.append(data_72_73[0,state_72_73s-1]*scl)
energy_73.append(data_71_74[0,state_71_73s-1]*scl)
energy_73.append(data_70_75[0,state_70_73s-1]*scl)
```

```python
num_n = [0,1,2,3]
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
start_n = 0
plt.plot(num_n[start_n:], energy_73[start_n:], '-o')
plt.xlabel("# of n manifolds either side")
plt.ylabel('Energy of 73s (GHz)')
plt.grid()
plt.title('All data points')

plt.subplot(1, 2, 2)
start_n = 1
plt.plot(num_n[start_n:], energy_73[start_n:], '-o')
plt.xlabel("# of n manifolds either side")
plt.ylabel('Energy of 73s (GHz)')
plt.grid()
plt.title('All expect first')
```

```python
num_n = [1,2,3]
plt.figure(figsize=(12,4))
plt.subplot(1, 2, 1)
start_n = 0
plt.plot(num_n[start_n:], np.abs(np.diff(np.array(energy_73[start_n:])))*10**3, '-o')
plt.xlabel("# of n manifolds either side")
plt.ylabel('d E_73s (MHz)')
plt.grid()
plt.title('All data points')

plt.subplot(1, 2, 2)
start_n = 1
plt.plot(num_n[start_n:], np.abs(np.diff(np.array(energy_73[start_n:])))*10**3, '-o')
plt.xlabel("# of n manifolds either side")
plt.ylabel('d E_73s (MHz)')
plt.grid()
plt.title('All expect first')
```

```python
print('Different from 2 to 3 either side: ', np.abs(np.diff(np.array(energy_73[start_n:])))[1]*10**3, ' (MHz)')
```

We're using 69 <= n <= 74 for 73s-70c transition. This provides 2 n manifolds either side. The plot suggests that adding an extra either side would increase the accuracy by around 0.863 MHz. Given that the current error is around 12MHz, it is unlikely to solve the problem. However this data was take at E=0.84 V/cm. It would be more consistent to use E=0.80 V/cm, which I am now currently running.
